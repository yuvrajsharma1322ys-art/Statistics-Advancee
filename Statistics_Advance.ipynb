**4. What is a binomial distribution, and how is it used in probability?**

- A binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials of a random experiment, where each trial has only two possible outcomes: success or failure.

-  ** Conditions for Binomial Distribution**

   - A random experiment follows a binomial distribution if:

   - The number of trials (n) is fixed

   - Each trial is independent

   - Each trial has only two outcomes (success/failure)

   - Probability of success (p) remains constant

   - Probability of failure (q = 1 − p)

- Uses of Binomial Distribution in Probability & Data Science

  - Modeling success/failure experiments

  - Quality control (defective vs non-defective items)

  - A/B testing in data science

  - Survey analysis (yes/no responses)

  - Medical trials (effective/not effective)

**6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
- The Central Limit Theorem (CLT) is one of the most important theorems in statistics. It states that the distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the original population’s distribution, provided the samples are independent and have a finite mean and variance.

- Why CLT Is Critical in Statistics

  - Justifies Use of Normal Distribution
  - Even if the original data is not normally distributed, CLT allows us to use normal-based methods for large samples.

- Foundation of Inferential Statistics
  - CLT is the basis for:

    - Hypothesis testing

    - Confidence intervals

    - Regression analysis

    - Simplifies Real-World Analysis Makes it possible to analyze complex and unknown population distributions.

- Essential in Data Science Used in:

  - A/B testing

  - Sampling and estimation

  - Predictive modeling

7. What is the significance of confidence intervals in statistical analysis?
- A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as mean or proportion) with a specified level of confidence (e.g., 95%).

- Why Confidence Intervals Are Important

   - Measure of Uncertainty
   - Confidence intervals show how precise an estimate is, rather than giving just a single point estimate.

- Better Than Point Estimates Alone
  - A point estimate (like sample mean) does not indicate reliability, but a confidence interval provides a range of plausible values.

- Widely Used in Statistical Inference: Confidence intervals are used in:

   - Hypothesis testing

   - Estimating population parameters

   - Comparing groups

- Decision Making
  -  Helps researchers and data scientists assess whether results are statistically meaningful and practically significant.

- Common in Data Science & Research
  - Used in A/B testing, surveys, medical studies, and machine learning model evaluation.

8. What is the concept of expected value in a probability distribution?
-  The expected value (EV) is the average or mean value of a random variable, calculated by weighting each possible outcome by its probability. It represents the long-run average result if a random experiment is repeated many times.

- Importance in Statistics and Data Science

  - Summarizes a Distribution: Gives a single value representing the center of the distribution.

  - Decision Making Under Uncertainty: Used in risk analysis, economics, and game theory.

  - Foundation for Other Concepts: Variance, standard deviation, and many statistical models depend on expected value.

  - Used in Machine Learning: Appears in loss functions, optimization, and probabilistic models.


9. Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normal Distribution (Mean=50, Std=5)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

10. You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.

```python
import numpy as np
from scipy import stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
sales = np.array(daily_sales)

# Sample size
n = len(sales)

# Mean and standard deviation
mean_sales = np.mean(sales)
std_sales = np.std(sales, ddof=1)

# Standard error
standard_error = std_sales / np.sqrt(n)

# Z-score for 95% confidence interval
z = 1.96

# Confidence interval
lower_bound = mean_sales - z * standard_error
upper_bound = mean_sales + z * standard_error

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```

- The 95% confidence interval gives a range within which we are 95% confident that the true average daily sales of the company lie. This helps management understand the overall sales trend and make informed business decisions.
